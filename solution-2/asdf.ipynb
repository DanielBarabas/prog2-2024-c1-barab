```python
df = pd.read_csv("input.csv")
query_df = pd.read_csv("query.csv")
```

```python
ls = []
sumls = []
for idx, row in query_df.iterrows():

        diffs = (df[["x", "y"]] - row).pow(2)
        sum = diffs.sum(axis=1)
        ls.append(diffs)
        sumls.append(sum)
```
